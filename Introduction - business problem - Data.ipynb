# Capstone Project - Cafes in Istanbul / Turkey


## Introduction / business problem

In a city of Istanbul, if someone is looking to open a cafe restaurant, **the question is**, where would you recommend that they open it? **The background of the problem** is that in order for a cafe to be profitable, there must be enough customers, and in order to have enough customers, it is not worth setting up a cafe in the immediate promixity of existing ones.

Let's also make sure that **audience is explicitly defined** to be the local restaurant entrepreuners in Istanbul and **they should care about this problem** because the location of the new café has a significant impact on the expected returns.
